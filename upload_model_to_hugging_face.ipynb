## 3. Configure Upload Settings

````python

HUGGINGFACE_USERNAME = "pshashid"
REPO_NAME_ENGLISH = "xlmr-toxicity-english"
REPO_NAME_MULTILINGUAL = "xlmr-toxicity-multilingual"

# Local file paths (these are from your training)
ENGLISH_MODEL_PATH = "/content/gdrive/MyDrive/Models/NLP/xlmr_large_global_v1_epoch2.bin"
MULTILINGUAL_MODEL_PATH = "/content/gdrive/MyDrive/Models/NLP/xlmr_multi_large_global_v1_epoch2.bin"

# Construct full repo IDs
repo_id_english = f"{HUGGINGFACE_USERNAME}/{REPO_NAME_ENGLISH}"
repo_id_multilingual = f"{HUGGINGFACE_USERNAME}/{REPO_NAME_MULTILINGUAL}"

print("UPLOAD CONFIGURATION")
print(f"\nUsername: {HUGGINGFACE_USERNAME}")
print(f"English model repo: {repo_id_english}")
print(f"Multilingual model repo: {repo_id_multilingual}")
print(f"\nModel files:")
print(f"  English: {ENGLISH_MODEL_PATH}")
print(f"  Multilingual: {MULTILINGUAL_MODEL_PATH}")
````

## 4. Check Model Files Exist

````python
import os

print("\nChecking model files...\n")

en_exists = os.path.exists(ENGLISH_MODEL_PATH)
multi_exists = os.path.exists(MULTILINGUAL_MODEL_PATH)

print(f"English model exists: {en_exists}")
if en_exists:
    size_gb = os.path.getsize(ENGLISH_MODEL_PATH) / (1024**3)
    print(f"  Size: {size_gb:.2f} GB")

print(f"\nMultilingual model exists: {multi_exists}")
if multi_exists:
    size_gb = os.path.getsize(MULTILINGUAL_MODEL_PATH) / (1024**3)
    print(f"  Size: {size_gb:.2f} GB")

if not (en_exists and multi_exists):
    print("\n  WARNING: One or more model files not found!")
    print("Make sure the paths are correct above.")
else:
    print("\n Both model files found!")
````

## 5. Create Repositories on Hub

````python
from huggingface_hub import create_repo

print("CREATING REPOSITORIES")

print(f"\n1. Creating English model repository...")
create_repo(
    repo_id=repo_id_english,
    private=False,  # Set to True if you want it private
    exist_ok=True   # Won't fail if repo already exists
)
print(f" Ready: https://huggingface.co/{repo_id_english}")

print(f"\n2. Creating Multilingual model repository...")
create_repo(
    repo_id=repo_id_multilingual,
    private=False,
    exist_ok=True
)
print(f" Ready: https://huggingface.co/{repo_id_multilingual}")

````

## 6. Upload Model Weights

````python
from huggingface_hub import upload_file
import time

print("UPLOADING MODEL WEIGHTS")
print("\n(This may take several minutes depending on file size)\n")

# Upload English model
print("1. Uploading English model...")
start_time = time.time()
try:
    upload_file(
        path_or_fileobj=ENGLISH_MODEL_PATH,
        path_in_repo="xlmr_toxicity_english.bin",
        repo_id=repo_id_english,
        commit_message="Upload trained English toxicity model (epoch 2)"
    )
    elapsed = time.time() - start_time
    print(f" Uploaded successfully in {elapsed:.1f} seconds")
    print(f"   URL: https://huggingface.co/{repo_id_english}/blob/main/pytorch_model.bin")
except Exception as e:
    print(f" Error: {e}")

# Upload Multilingual model
print("\n2. Uploading Multilingual model...")
start_time = time.time()
try:
    upload_file(
        path_or_fileobj=MULTILINGUAL_MODEL_PATH,
        path_in_repo="xlmr_toxicity_multilingual.bin",
        repo_id=repo_id_multilingual,
        commit_message="Upload trained multilingual toxicity model (epoch 2)"
    )
    elapsed = time.time() - start_time
    print(f" Uploaded successfully in {elapsed:.1f} seconds")
    print(f"   URL: https://huggingface.co/{repo_id_multilingual}/blob/main/pytorch_model.bin")
except Exception as e:
    print(f" Error: {e}")

````

## 7. Upload Model Cards (README.md)

````python
from huggingface_hub import upload_file
import tempfile
import os

print("UPLOADING MODEL CARDS (README.md)")

# English model card
english_readme = """---
language:
- en
license: mit
library_name: transformers
tags:
- toxicity-detection
- xlm-roberta
---

# XLM-RoBERTa Toxicity Detection (English)

## Overview
This model detects toxic comments in English and classifies the type of toxicity intent.

## Model Details
- **Architecture**: XLM-RoBERTa-Large (384M parameters)
- **Task**: Binary toxicity classification + Multi-label intent classification
- **Intents**: severe_toxic, obscene, threat, insult, identity_hate
- **Training Data**: Jigsaw Toxic Comment Classification Challenge (161k samples)

## Performance
- **Accuracy**: 88%
- **F1-Score**: 0.88
- **AUC-ROC**: 0.9486

## Usage
```python
import torch
from transformers import AutoTokenizer
from huggingface_hub import hf_hub_download
from model_utils import MultiTaskXLMR

# Load model
model_path = hf_hub_download("{repo_id_english}", "pytorch_model.bin")
model = MultiTaskXLMR(model_name="xlm-roberta-large", num_intents=5)
model.load_state_dict(torch.load(model_path))
model.eval()

# Inference
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-large")
text = "You are awesome!"
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

with torch.no_grad():
    tox_logits, intent_logits = model(inputs['input_ids'], inputs['attention_mask'])
    tox_prob = torch.sigmoid(tox_logits).item()
    print(f"Toxicity: {tox_prob:.4f}")
```

## License
MIT
"""

# Multilingual model card
multilingual_readme = """---
language:
- de
- es
- fr
- it
- ru
- zh
- ja
- ar
- he
- am
- tt
- uk
- hi
license: mit
library_name: transformers
tags:
- toxicity-detection
- multilingual
- xlm-roberta
---

# XLM-RoBERTa Toxicity Detection (Multilingual)

## Overview
This model detects toxic comments across **14 languages** using a single, unified model.

## Supported Languages
German, Spanish, French, Italian, Russian, Chinese, Japanese, Arabic, Hebrew, Amharic, Tatar, Ukrainian, Hindi

## Model Details
- **Architecture**: XLM-RoBERTa-Large (384M parameters)
- **Task**: Binary toxicity classification
- **Training Data**: TextDetox Multilingual Toxicity Dataset

## Per-Language Performance (F1-Score)
- French: 0.9931
- Russian: 0.9877
- Hindi: 0.9403
- Ukrainian: 0.9401
- Japanese: 0.8917
- Spanish: 0.8846
- Italian: 0.8834
- German: 0.8823
- Tatar: 0.8681
- Arabic: 0.8310
- Amharic: 0.8267
- Chinese: 0.8150
- Hebrew: 0.7829

## Usage
```python
import torch
from transformers import AutoTokenizer
from huggingface_hub import hf_hub_download
from model_utils import MultiTaskXLMR

# Load model
model_path = hf_hub_download("{repo_id_multilingual}", "pytorch_model.bin")
model = MultiTaskXLMR(model_name="xlm-roberta-large", num_intents=5)
model.load_state_dict(torch.load(model_path))
model.eval()

# Inference in any supported language
tokenizer = AutoTokenizer.from_pretrained("xlm-roberta-large")
text = "Tu es un idiot!"  # French
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)

with torch.no_grad():
    tox_logits, _ = model(inputs['input_ids'], inputs['attention_mask'])
    tox_prob = torch.sigmoid(tox_logits).item()
    print(f"Toxicity: {tox_prob:.4f}")
```

## License
MIT
"""

# Save README files temporarily and upload
print("\n1. Uploading English model card...")
with tempfile.NamedTemporaryFile(mode='w', suffix='.md', delete=False) as f:
    f.write(english_readme)
    temp_en = f.name

try:
    upload_file(
        path_or_fileobj=temp_en,
        path_in_repo="README.md",
        repo_id=repo_id_english,
        commit_message="Add model card"
    )
    print("  Uploaded")
except Exception as e:
    print(f" Error: {e}")
finally:
    os.remove(temp_en)

print("\n2. Uploading Multilingual model card...")
with tempfile.NamedTemporaryFile(mode='w', suffix='.md', delete=False) as f:
    f.write(multilingual_readme)
    temp_multi = f.name

try:
    upload_file(
        path_or_fileobj=temp_multi,
        path_in_repo="README.md",
        repo_id=repo_id_multilingual,
        commit_message="Add model card"
    )
    print(" Uploaded")
except Exception as e:
    print(f" Error: {e}")
finally:
    os.remove(temp_multi)
````
